### Question 1
Imagine we have a data frame called `data`, with a `type` column. Which one works and why?\
Function 1:
```
group_and_tally <- function(df, column){
    df %>% group_by({{ column }}) %>% tally();
}
group_and_tally(data, type);
```

Function 2:
```
group_and_tally <- function(df, column){
    df %>% group_by(column) %>% tally();
}
group_and_tally(data, type);
```

Function 1 works and function 2 does not work. The {{}} in the first function unquote the argument within the brackets. This will inject the different column names into the groupby() function. "column" without the brackets in the second function means literal 'column'. Since there are no column name called 'column' the function will not group the dataframe by the column values.

### Question 8
Notice something interesting about the names in our data set. Fix the name formatting and print the first 10 observations.

````r
# the names have a suffix of 3 numbers
remove_numbers <- function(x) {
  gsub("[0-9]+$", "", x)
}

joined[c('FIRST','LAST')] <- lapply(joined[c('FIRST','LAST')], remove_numbers)
head(joined, 10)
````

### Question 9
Using a for statement to loop through the categorical variables (excluding name and ID), print the counts of each unique value in descending order, using the mdpre() function for formatting.

````r
cat_cols <- setdiff(names(joined), c("ID", "FIRST", "LAST"))

# Loop through categorical columns and print counts
for (col in cat_cols) {
  counts <- sort(table(joined[[col]]), decreasing = TRUE)
  cat(col, ":\n")
  mdpre(counts)
}
````

### Question 10
If you see any weird values, get rid of the ones that don't make sense, and combine the ones that are formatted wrong. Don't forget ot check the dates! Print the new tables for categorical values, and print the date ranges.

````r
#marital: drop Fine and male
# Drop rows where MARITAL is 'Fine'
joined <- joined[joined$MARITAL != "Fine", ]
joined <- joined[joined$MARITAL != "male", ]
````

````r
#race: change asiann into asian
joined$RACE[joined$RACE == "asiann"] <- "asian"
````

````r
#ethnicity: spelling of hispanic
joined$ETHNICITY[joined$ETHNICITY == "nonhispani"] <- "nonhispanic"
joined$ETHNICITY[joined$ETHNICITY == "hispani"] <- "hispanic"
````

````r
#gender
joined$GENDER[joined$GENDER == "female"] <- "F"
joined$GENDER[joined$GENDER == "Female"] <- "F"
joined$GENDER[joined$GENDER == "Male"] <- "M"
````

````r
ref_date <- as.Date("2020-01-01")

#for birth dates and death dates after today, minus 100 years
joined$BIRTHDATE <- ifelse(
  joined$BIRTHDATE > ref_date,
  make_date(year = year(joined$BIRTHDATE) - 100,
            month = month(joined$BIRTHDATE),
            day = day(joined$BIRTHDATE)),
  joined$BIRTHDATE
)
joined$BIRTHDATE <- as.Date(joined$BIRTHDATE)

joined$DEATHDATE <- ifelse(
  joined$DEATHDATE > ref_date,
  make_date(year = year(joined$DEATHDATE) - 100,
            month = month(joined$DEATHDATE),
            day = day(joined$DEATHDATE)),
  joined$DEATHDATE
)

# Convert back to Date
joined$DEATHDATE <- as.Date(joined$DEATHDATE)
````

````r
#drop na
joined <- joined %>%
  filter(if_all(-DEATHDATE, ~ !is.na(.)))
````

````r
cat_cols <- setdiff(names(joined), c("ID", "FIRST", "LAST"))

# Loop through categorical columns and print counts
for (col in cat_cols) {
  counts <- sort(table(joined[[col]]), decreasing = TRUE)
  cat(col, ":\n")
  mdpre(counts)
}

#date range
min_b <- min(joined$BIRTHDATE, na.rm = TRUE)
max_b <- max(joined$BIRTHDATE, na.rm = TRUE)

min_d <- min(joined$DEATHDATE, na.rm = TRUE)
max_d <- max(joined$DEATHDATE, na.rm = TRUE)

min <- min(min_b, min_d)
max <- max(max_b, max_d)

print(paste("Date range is", min, "to", max))
````
