## Question 1
#### What is the holy trinity of Git? (Just name the parts.)

## Question 2
#### Explain how patches relate to the idea of committing a change to the history of your code?

Each commit is a patch that transform the code from old to new state. This includes changes that are added to the file and changes that are deleted from the file. 

## Question 3
#### What is the difference between a branch and a fork?

A branch is a pointer to a series of commit in the repository. Branches are created to add features or make changes to the main branch without affecting the main branch. A fork is a copy of the repo under the local device. 

## Question 4
#### How are diff and patch related? What’s the difference?

Git diff shows the difference between two files and patch is the program that takes the diff and transform he first file into the second one.
